```python
# set base path
base_path = "../"

# set data path for images
image_path = "data/testset/images/"

# set path for weights
weights_path = "data/train_final_checkpoints/"

# set path for predictions
predict_path = 'data/testset/predictions/'

# dataset sizes (for looping)
sizes = ["small","medium","large","small_transform","medium_transform","large_transform"]
```

```python
# set model parameters
model_params = {"input_height": 800,
                "input_width": 800,
                "n_classes" : 2}
```

```python
# set models
Vanilla8 = fcn_8(**model_params)
Vanilla32 = fcn_32(**model_params)
VGG8 = fcn_8_vgg(**model_params)
VGG32 = fcn_32_vgg(**model_params)
Resnet8 = fcn_8_resnet50(**model_params)
Resnet32 = fcn_32_resnet50(**model_params)
```

```python
# create model dict
models = {"Vanilla8": Vanilla8,
         "Vanilla32": Vanilla32,
         "VGG8": VGG8,
         "VGG32": VGG32,
         "Restnet8": Resnet8,
         "Restnet32": Resnet32}
```

```python
# create folder tree for predictions
for model in models.keys():
    for size in sizes:
        size = "/"+size
        os.makedirs(predict_path+model+size, exist_ok=True)
```

```python
# get list of image files and create input path
im_files = os.listdir(image_path)
input_path = [image_path+f for f in os.listdir(image_path)]
```

```python
im_files
```

```python
# model predictions
for model_name, model in models.items():
    for size in sizes:
        
        # load weights
        weightfile = [re.sub(".index","",f) for f in os.listdir(weights_path+model_name+"/"+size+"/") if f.endswith(".index")]
        print(weightfile)
        model.load_weights(filepath = weights_path+model_name+"/"+size+"/"+weightfile[0])
        
        # create output path
        output_path = [predict_path+model_name+"/"+size+"/"+f for f in im_files]
        
        for f_in, f_out in zip(input_path, output_path):
            
            model.predict_segmentation(
            inp= f_in,
            out_fname= f_out
            )
        
```
